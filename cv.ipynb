```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 对于一个属性，共m方提供的数据包含该属性，每一方数据被篡改的概率为pi，篡改成本为篡改失败的概率

def C1(m, p):
    return (1-p)**m

def C2(m, p):
    return 1-p**m
```

```python
# 传统区块链溯源方案 vs 多方交叉验证溯源方案
cost1 = C1(np.array([1, 3, 5, 7, 9]), 0.5)
cost2 = C2(np.array([1, 3, 5, 7, 9]), 0.5)
cost1, cost2

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.plot([1,3,5,7,9], cost1, marker='o', label='传统方案')
plt.plot([1,3,5,7,9], cost2, marker='o', label='交叉验证')
plt.legend()
plt.xlabel('参与者数量')
plt.ylabel('篡改成本')
```
